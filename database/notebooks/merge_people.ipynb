```python
import json
from rich import print, inspect
from pathlib import Path
```

```python
people_grouped_file = Path("../in_progress/people_grouped.json")
people_file = Path("../in_progress/people_results.json")
folder_path = Path("../in_progress/pass2_results")
# folder_path = folder_path.resolve()
```

Check for inconsistencies from API parsing

```python

parsed_people = []
check_inconsistencies = {}
inconsistent = []

with open(people_file, "r") as f:
    parsed_people = json.load(f)

    for entry in parsed_people:
        unified_id = entry["unified_id"]
        family_name = entry.get("family_name")
        given_names = entry.get("given_names")
        names = (family_name, given_names)


        if unified_id not in check_inconsistencies:
            check_inconsistencies[unified_id] = {names}
        else:
            check_inconsistencies[unified_id].add(names)

        count = len(check_inconsistencies[unified_id])
        inconsistent_count = 0
        if count > 1:
            inconsistent_count += 1
            inconsistent.append(unified_id)
inconsistent_people = [uid for uid, name_set in check_inconsistencies.items() if len(name_set) > 1]
print(f"Total unique people with inconsistencies: {len(inconsistent_people)}")
```
