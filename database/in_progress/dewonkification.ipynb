```python
from rich import print as rprint
from rich import inspect
import json
import re
```

```python
with open("people_grouped.json", "r") as f:
    people = json.load(f)
```

```python
wonky = {}
corrected = {}
```

```python
delimiters = [" und ", " u.", ",", ";"]
particles = ["di", "vom", "de", "da", "de la", "zu", "von", "van", "v."]
org_keywords = ["stiftung", "archiv", "gesellschaft", "forum", "gemeinde", "sammlung", "institut", "museum", "kultur", "kuratorium", "verlag", "ministerium", "akademie", "trust", "verein", "organisation", "vereinigung", "universit√§t", "bibliothek"]
```

```python
for name, entries in people:
    original_name = name

    for person in entries:
        if (
            person["family_name"] is None and
            person["given_names"] is None and
            person["name_particles"] is None and
            person["is_organisation"] is False
        ) and any(delimiter in original_name for delimiter in delimiters):
            wonky[name] = entries
```

```python
for name_key, entries in wonky.items():
    for entry in entries:
        display_name = entry["display_name"]
        single_name = entry["single_name"]

        text2split = display_name or single_name

        if not text2split or text2split == "null":
            print(f"Nothing to split for {name_key}!")
            continue

        if " und " in text2split:
            has_und = text2split.split(" und ")
            # print(has_und)
            for name in has_und:
                if "," in name:
                    has_comma = name.split(", ")
                    print(has_comma)
```
