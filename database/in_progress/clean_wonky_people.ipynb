```python
review = {}
for key in wonky_data.keys():
    if key == "oops, something went wrong":
        if key not in review:
            review[key] = wonky_data[key]
        else:
            review[key].append(wonky_data[key])
        continue

    particle = None
    names = key.strip().split()
    if len(names) == 1:
        if key not in review:
            review[key] = wonky_data[key]
        else:
            review[key].append(wonky_data[key])
        continue

    if len(names) == 2:
        assign_names = names
        given = assign_names[0].title()
        family = assign_names[1].title()
        # first_last.append(key)
    if len(names) >= 3:
        longer_names = names
        if longer_names[1]in ["v.", "von", "van"]:
            given = longer_names[0].title()
            particle = longer_names[1]
            family = longer_names[2].title()
        else:
            given = longer_names[0].title() + " " + longer_names[1].title()
            family = longer_names[2].title()

    for entry in wonky_data[key]:
        entry["family_name"] = family
        entry["given_names"] = given
        entry["name_particles"] = particle
print(len(review))
# inspect(review)
```
