```python
import json
import sys
from rich import print as rprint
from collections import defaultdict

from pathlib import Path
from itertools import islice

nb_dir = Path.cwd()

project_root = nb_dir.parent.parent

sys.path.insert(0, str(project_root))
```

```python
people_records_prepped_file = Path(project_root / "data/people/people_records_prepped.json")
books2people_prepped_file_old = Path(project_root / "data/people/books2people_prepped_old.json")
books2people_prepped_file = Path(project_root / "data/people/books2people_prepped.json")
folder_matched = Path(project_root / "data/matched")
folder_validated = Path(project_root / "data/validated")
```

```python
b2p_new = set()
deduplicated = []

with open(books2people_prepped_file_old, "r") as f:
    b2p_old = json.load(f)

for entry in b2p_old:
    entry_str = json.dumps(entry, sort_keys=True)

    if entry_str not in b2p_new:
        b2p_new.add(entry_str)
        deduplicated.append(entry)

with open(books2people_prepped_file, "w") as f:
    json.dump(deduplicated, f, ensure_ascii=False, indent=2)
```
