```python
import json
import sys
from rich import print as rprint
from pathlib import Path

from itertools import batched

nb_dir = Path.cwd()

project_root = nb_dir.parent.parent

sys.path.insert(0, str(project_root))
```

```python
people_new_file = Path(project_root / "data_reload/people/new_people_data.json")
people_old_file = Path(project_root / "data_reload/people/people_clean_old_id.json")

with open(people_new_file, "r") as f:
   people_new = json.load(f)
with open(people_old_file, "r") as f:
   people_old = json.load(f)

people_all = people_new + people_old

rprint(f"new: {len(people_new)}")
rprint(f"old: {len(people_old)}")
rprint(f"all: {len(people_all)}")

people_file = Path(project_root / "data_reload/people/people_file.json")
with open(people_file, "w") as f:
    json.dump(people_all, f, ensure_ascii=False, indent=2)
```
