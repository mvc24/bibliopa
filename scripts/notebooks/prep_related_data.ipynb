```python
import json
import sys
from rich import print as rprint
from pathlib import Path
import re

nb_dir = Path.cwd()

project_root = nb_dir.parent.parent

sys.path.insert(0, str(project_root))
```

```python
book_file = Path(project_root / "data_reload/books/books_complete.json")
book_db_file = Path(project_root / "data/from db/books_db.json")

with open(book_file, "r") as f:
   books = json.load(f)

with open(book_db_file, "r") as f:
   db_books = json.load(f)

book_id_dict = {v["composite_id"]: v["book_id"] for v in db_books}
# rprint(dict(list(book_id_dict.items())[:3]))

admin = [a[0]["admin_data"] for k, a in books.items()]

admin_data = []
not_found = []

# for entry in admin[:10]:
for entry in admin:
    composite_id = entry["composite_id"]
    if not composite_id in book_id_dict:
        not_found.append(entry)
        continue
    book_id = book_id_dict[composite_id]

    admin_data.append({
       "book_id": book_id,
       **entry
    })

rprint(f"not found: {len(not_found)}")

# rprint(admin_data)
# rprint(f"processed: {len(admin_data)}")

# with open("admin_data.json", "w") as f:
#     json.dump(admin_data, f, ensure_ascii=False, indent=2)
```

```python
prices_data = []
not_found_prices = []

for composite_id, entry_list in books.items():
    if composite_id not in book_id_dict:
        not_found_prices.append(composite_id)
        continue
    book_id = book_id_dict[composite_id]
    price = entry_list[0]["price_data"]

    prices_data.append({
        "book_id": book_id,
        **price
    })

rprint(f"not found: {len(not_found_prices)}")
rprint(f"processed: {len(prices_data)}")

with open("prices_data.json", "w") as f:
    json.dump(prices_data, f, ensure_ascii=False, indent=2)
```

```python
volumes_data = []
not_found_volumes = []

for composite_id, entry_list in books.items():
    if composite_id not in book_id_dict:
        not_found_volumes.append(composite_id)
        continue
    book_id = book_id_dict[composite_id]
    volumes = entry_list[0]["volumes_data"]["volumes"] or []

    for volume in volumes:
        volumes_data.append({
            "book_id": book_id,
            **volume
        })

rprint(f"not found: {len(not_found_volumes)}")
rprint(f"volumes processed: {len(volumes_data)}")

with open("volumes_data.json", "w") as f:
    json.dump(volumes_data, f, ensure_ascii=False, indent=2)
```
