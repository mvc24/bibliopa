```python
import json
import sys
from rich import print as rprint
from pathlib import Path
import re

nb_dir = Path.cwd()

project_root = nb_dir.parent.parent

sys.path.insert(0, str(project_root))
```

```python
people_clean_folder = Path(project_root / "data/api_people/clean_results")
people_clean_file = Path(project_root / "data/people/people_clean.json")

nopes_clean_folder = Path(project_root / "data/api_people/nopes_results")
nopes_clean_file = Path(project_root / "data/people/ nopes_clean.json")
```

```python
people_clean = []
person_count = 0

for file in people_clean_folder.iterdir():

    with open(file, "r") as f:
       people = json.load(f)

       for person in people:
           person_count += 1
           people_clean.append(person)
# rprint(person_count)
# with open(people_clean_file, "w") as f:
#     json.dump(people_clean, f, ensure_ascii=False, indent=2)
```

```python
nopes_clean = []
nopes_count = 0

for file in nopes_clean_folder.iterdir():

    with open(file, "r") as f:
        people_new = json.load(f)

    for person in people_new:
        nopes_count += 1
        nopes_clean.append(person)

rprint(nopes_count)
with open(nopes_clean_file, "w") as f:
    json.dump(nopes_clean, f, ensure_ascii=False, indent=2)
```
