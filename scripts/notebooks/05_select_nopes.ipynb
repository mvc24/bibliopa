```python
found_nopes_file = Path(project_root / "scripts/notebooks/found_nopes_file.json")
with open(found_nopes_file, "r") as f:
   f_n = json.load(f)
rprint(f"entries: {len(f_n)}")

db_people_file = Path(project_root / "data_reload/db_files/people.json")
with open(db_people_file, "r") as f:
   people = json.load(f)
people_dict = {i["unified_id"]: i for i in people}
rprint(dict(list(people_dict.items())[:2]))

rprint(dict(list(f_n.items())[:3]))
```

```python
matched = {}
unmatched = {}

for cid, entries in f_n.items():
    for entry in entries:
        uid = entry["unified_id"]
        display_norm = entry["display_norm"]
        person = people_dict.get(uid)
        if person:
            matched.setdefault(cid, []).append(
                {"unified_id": uid, "person_id": person["person_id"], "display_norm": display_norm}
            )
        else:
            unmatched.setdefault(cid, []).append(
                {"unified_id": uid, "display_norm": display_norm}
            )

rprint(f"matched cids: {len(matched)}  |  unmatched cids: {len(unmatched)}")
rprint(dict(list(matched.items())[:2]))
rprint(dict(list(unmatched.items())[:2]))
with open("matched_from_nope.json", "w") as f:
    json.dump(matched, f, ensure_ascii=False, indent=2)
with open("unmatched_from_nope.json", "w") as f:
    json.dump(unmatched, f, ensure_ascii=False, indent=2)
```
