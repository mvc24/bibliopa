```python
people_folder = Path(project_root / "data/people")
report = []

for file in people_folder.rglob("*.json"):
   rel_path = file.parent.relative_to(project_root.parent)
   filename = file.name
   path = str(rel_path / file.name)


   with open(file, "r") as f:
      data = json.load(f)
      # report_s = f"""
      # {filename}
      # {len(data)} entries
      # Last modified: {datetime.fromtimestamp(file.stat().st_mtime)}
      # """
      if isinstance(data, list):
         first_entry = data[0]
         structure = "list"
      else:
         first_entry = next(iter(data.items()))
         structure = "dict"

      report.append({
         "file": filename,
         "path": path,
         "entries": len(data),
         "modified": datetime.fromtimestamp(file.stat().st_mtime).strftime("%Y-%m-%d %H:%M"),
         "timestamp": file.stat().st_mtime,
         "structure": structure,
         "first_entry": first_entry
      })
      report_sorted = sorted(report, key=lambda x: x["timestamp"], reverse=True)
      report_longest_file = sorted(report, key=lambda x: x["entries"], reverse=True)


with open("people_report.json", "w") as f:
    json.dump(report_sorted, f, ensure_ascii=False, indent=2)
with open("people_by_entries.json", "w") as f:
    json.dump(report_longest_file, f, ensure_ascii=False, indent=2)


with open("people_files.csv", mode="w") as csv_file:
   fieldnames = list(report_sorted[0].keys())
   writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
   writer.writeheader()
   writer.writerows(report_sorted)


   # with open("people_files.txt", "a") as f:
   #    f.write(report)
```

```python
books_prepped_folder = Path(project_root / "data_reload/books/prepped")

total_books = 0
report = []

for file in books_prepped_folder.iterdir():

    filename = file.name
    with open(file, "r") as f:
        books = json.load(f)
    book_count = len(books)
    report.append({
        "filename": filename,
        "book_count": book_count
    })
    total_books += book_count

# rprint(f"total_books: {total_books}")
with open("book_stats.csv", "w") as csv_file:
    fieldnames = list(report[0].keys())
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(report)
```
