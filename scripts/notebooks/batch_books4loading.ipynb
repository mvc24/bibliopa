```python
import json
import sys
from rich import print as rprint
from pathlib import Path
import re
from itertools import batched

nb_dir = Path.cwd()

project_root = nb_dir.parent.parent

sys.path.insert(0, str(project_root))
```

```python
books_folder = Path(project_root / "data_reload/books/prepped")
batched_folder = Path(project_root / "data_reload/books/loading_batches")

total_books = 0
all_books = {}

for file in books_folder.iterdir():
    with open(file, "r") as f:
       books = json.load(f)
    book_count = len(books)
    total_books += book_count
    all_books.update(books)

# rprint(f"total after loop: {total_books}")
# rprint(dict(list(all_books.items())[:2]))
```

```python
i = 0

for batch in batched(all_books.items(), 100):
    book_batch = dict(batch)
    i += 1
    batch_id = str(i).zfill(2)
    filename = "books4loading_" + batch_id + ".json"
    file_path = batched_folder / filename
    # rprint(file_path)
    with open(file_path, "w") as f:
        json.dump(book_batch, f, ensure_ascii=False, indent=2)
    rprint(f"wrote & saved {filename}")
```
