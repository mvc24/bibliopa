```python
with open(multiples_file, "r") as f:
    multiples = json.load(f)

SEPARATORS = re.compile(r"\s*;\s*|\s+/\s+|\s+und\s+")
ETC_SUFFIX = re.compile(r"\s+u\.a\.\s*$")

multiples_dict = {}

for names, data in multiples.items():
    ids = data["person_ids"]
    entries = data["entries"]

    for i, id in enumerate(ids):
        sorting_new = i + 1
        rprint(f"i, id: {i, id}")

        for entry in entries:
            display_name = entry["display_name"]
            cleaned = ETC_SUFFIX.sub("", display_name).strip()
            parts = [p.strip() for p in SEPARATORS.split(cleaned) if p.strip()]
            indexed_name = parts[i]

            if id not in multiples_dict:
                multiples_dict[id] = [{
                    **entry,
                    "display_name": indexed_name,
                    "sort_order": sorting_new
                }]
            else:
                multiples_dict[id].append({
                    **entry,
                    "display_name": indexed_name,
                    "sort_order": sorting_new
                })
rprint(multiples_dict)

# with open("multiples_matched.json", "w") as f:
#     json.dump(multiples_dict, f, ensure_ascii=False, indent=2)

with open("multiples_matched.json", "w") as f:
    json.dump(multiples_dict, f, ensure_ascii=False, indent=2)



            # rprint(f"display: {display_name}")
            # rprint(f"cleaned: {cleaned}")
            # rprint(f"parts: {parts}")
            # rprint(f"indexed name: {parts[i]}")
        # multiples_dict[id] = [{
        #     **entries,
        #     "display_name": display_name_new,
        #     "sort_order": sorting_new,
        #     "last": last_name_new,
        #     etc...
        # }]




    # for ids, entry in data.items():
    #     rprint(ids, entry)
```
