```python
import json
import sys
from rich import print as rprint
from pprint import pp
from pathlib import Path
import re

nb_dir = Path.cwd()

project_root = nb_dir.parent.parent

sys.path.insert(0, str(project_root))
```

```python
new_people_file = Path(project_root / "data/people/prepping4loading/NEW_people_complete_checked.json")

with open(new_people_file, "r") as f:
   people = json.load(f)
```

```python
people_data = []
b2p_data = {}
b2p_list = []

for unified_id, person in people.items():
# for unified_id, person in list(people.items())[:2]:
    unified_id = person["unified_id"]
    family_name = person["family_name"]
    given_names = person["given_names"]
    name_prefix = person["name_prefix"]
    name_particles = person["name_particles"]
    name_suffix = person["name_suffix"]
    single_name = person["single_name"]
    is_organisation = person["is_organisation"]

    b2p_data[unified_id] = {
        "unified_id": unified_id,
        "family_name": family_name,
        "given_names": given_names,
        "name_prefix": name_prefix,
        "name_particles": name_particles,
        "name_suffix": name_suffix,
        "single_name": single_name,
        "is_organisation": is_organisation,
        "entries": []
    }
    people_data.append({
        "unified_id": person["unified_id"],
        "family_name": person["family_name"],
        "given_names": person["given_names"],
        "name_prefix": person["name_prefix"],
        "name_particles": person["name_particles"],
        "name_suffix": person["name_suffix"],
        "single_name": person["single_name"],
        "is_organisation": person["is_organisation"]
    })
    entries = person["entries"]
# rprint(people_data)
    for entry in entries:
        composite_id = entry["composite_id"]
        display_name = entry["display_name"]
        sort_order = entry["sort_order"]
        is_author = entry["is_author"]
        is_editor = entry["is_editor"]
        is_contributor = entry["is_contributor"]
        is_translator = entry["is_translator"]

        b2p_data[unified_id]["entries"].append({
            "display_name": display_name,
            "composite_id": composite_id,
            "sort_order": sort_order,
            "is_author": is_author,
            "is_editor": is_editor,
            "is_contributor": is_contributor,
            "is_translator": is_translator,
        })

        b2p_list.append({
            "display_name": display_name,
            "composite_id": composite_id,
            "sort_order": sort_order,
            "is_author": is_author,
            "is_editor": is_editor,
            "is_contributor": is_contributor,
            "is_translator": is_translator,
        })
        break

# rprint(f"b2p after: {b2p_data}")
# rprint(f"b2p list after: {b2p_list}")

prep_folder_path = Path(project_root / "data/people/prepping4loading")
new_people_file = "new_people_data.json"
people_path = prep_folder_path / new_people_file
with open(people_path, "w") as f:
    json.dump(people_data, f, ensure_ascii=False, indent=2)


b2p_no_ids_list = "b2p_no_ids_list.json"
b2p_no_ids_data = "b2p_no_ids_data.json"
```
