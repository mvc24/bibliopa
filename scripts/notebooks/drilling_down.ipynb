```python
matched = {}
singles = {}
probably_new = {}

for name, entry in list(checked.items())[:2]:
    points = entry["points"]
    info = entry["info"]
    composite_id = info["composite_id"]
    matches = entry["matches"]
    person_id = info["person_id"]
    correct_match = [m for m in matches if m[2]["person_id"] == person_id]
    rprint(f"match found {correct_match}")

    if person_id:
        if name in matched:
            matched[name].append({
                "composite_id": composite_id.extend(composite_id),
            })
        else:
            matched[name] = {
                "correct_match": correct_match,
                "person_id": person_id,
                "match_found": True,
                "composite_id": composite_id.extend(composite_id),
                "points": points
            }

# rprint(name, entry)
```
