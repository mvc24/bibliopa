```python
import json
import sys
from rich import print as rprint
from pathlib import Path
import re
from datetime import datetime

nb_dir = Path.cwd()
project_root = nb_dir.parent.parent
sys.path.insert(0, str(project_root))
```

```python
found_in_file = Path(project_root / "scripts/notebooks/found_in.json")
counted_files = Path(project_root / "scripts/notebooks/cids_per_file.json")
with open(found_in_file, "r") as f:
   cids_in_files = json.load(f)
with open(counted_files, "r") as f:
   files = json.load(f)

# rprint(files.keys())
nopes_files = [f for f in files if "nopes_results" in f]
# rprint(nopes_files)

nopes_results = []

for file in nopes_files:
    path = project_root / file
    with open(path, "r") as f:
        entries = json.load(f)
    nopes_results.extend(entries)

rprint(f"count: {len(nopes_results)}")
rprint(nopes_results[:1])
```
